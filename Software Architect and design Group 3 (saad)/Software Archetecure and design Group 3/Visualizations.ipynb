```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data from the Excel file
file_path = 'results.xlsx'
df = pd.read_excel(file_path)

# Set seaborn style for better visual aesthetics
sns.set(style='whitegrid')

# 1. Bar plot for Lines of Code (LOC) per project
plt.figure(figsize=(10, 6))
sns.barplot(x='Project', y='Lines of Code', data=df, palette='Blues_d')
plt.xticks(rotation=90)
plt.title('Lines of Code (LOC) per Project')
plt.tight_layout()
plt.show()

# 2. Bar plot for Cyclomatic Complexity per project
plt.figure(figsize=(10, 6))
sns.barplot(x='Project', y='Cyclometric Complexity', data=df, palette='Greens_d')
plt.xticks(rotation=90)
plt.title('Cyclomatic Complexity per Project')
plt.tight_layout()
plt.show()

# 3. Bar plot for Pylint Scores per project
plt.figure(figsize=(10, 6))
sns.barplot(x='Project', y='Pylint_scores', data=df, palette='Oranges_d')
plt.xticks(rotation=90)
plt.title('Pylint Scores per Project')
plt.tight_layout()
plt.show()

# 4. Scatter plot showing relationship between Lines of Code and Cyclomatic Complexity
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Lines of Code', y='Cyclometric Complexity', data=df, hue='Project', palette='Set2', s=100)
plt.title('Lines of Code vs Cyclomatic Complexity')
plt.tight_layout()
plt.show()

# 5. Scatter plot showing relationship between Cyclomatic Complexity and Pylint Scores
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Cyclometric Complexity', y='Pylint_scores', data=df, hue='Project', palette='Set1', s=100)
plt.title('Cyclomatic Complexity vs Pylint Scores')
plt.tight_layout()
plt.show()
```
